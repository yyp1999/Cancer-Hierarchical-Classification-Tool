```python
import pandas as pd
import os 
import joblib
import numpy as np
```

```python
#load models
f = open('markers/aftertukeybotuta.txt', 'r')
f_firstlayer = eval(f.read())
m_firstlayer=pd.read_csv('ref_mean/SUMmean.csv',index_col=0,header=None)
firstlayer_path='rfc_models/firstlayer.model'
firstlayer_model = joblib.load(firstlayer_path)
f = open('markers/group2_tubo.txt', 'r')
f_group2 = eval(f.read())
group2_path='rfc_models/group2.model'
group2_model = joblib.load(group2_path)
f = open('markers/group4_tubo.txt', 'r')
f_group4 = eval(f.read())
group4_path='rfc_models/group4.model'
group4_model = joblib.load(group4_path)
f = open('markers/group5_tubo.txt', 'r')
f_group5 = eval(f.read())
group5_path='rfc_models/group5.model'
group5_model = joblib.load(group5_path)
f = open('markers/group6_tubo.txt', 'r')
f_group6 = eval(f.read())
group6_path='rfc_models/group6.model'
group6_model = joblib.load(group6_path)
f = open('markers/group7_tubo.txt', 'r')
f_group7 = eval(f.read())
group7_path='rfc_models/group7.model'
group7_model = joblib.load(group7_path)
f = open('markers/group8_tubo.txt', 'r')
f_group8 = eval(f.read())
group8_path='rfc_models/group8.model'
group8_model = joblib.load(group8_path)
f = open('markers/group10_tubo.txt', 'r')
f_group10 = eval(f.read())
group10_path='rfc_models/group10.model'
group10_model = joblib.load(group10_path)
f = open('markers/group11_tubo.txt', 'r')
f_group11 = eval(f.read())
group11_path='rfc_models/group11.model'
group11_model = joblib.load(group11_path)
f = open('markers/group12_tubo.txt', 'r')
f_group12 = eval(f.read())
group12_path='rfc_models/group12.model'
group12_model = joblib.load(group12_path)
f.close()
```

```python
def PredictForsample(inputs):
    test = list(inputs.index)
    l = len(test)
    for i in range(l):
        test[i]=test[i].split('.')[0]
    res2 = []
    for i in range(l):
        testtemp = firstlayer_model.predict([inputs.iloc[i][f_firstlayer]])[0]
        if testtemp == 'group_1':
            res2.append('OV')
        elif testtemp == 'group_2':
            try:

                res2.append(group2_model.predict([inputs.iloc[i][f_group2]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_3':
            res2.append('THYM')
        elif testtemp == 'group_4':
            try:
                res2.append(group4_model.predict([inputs.iloc[i][f_group4]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_5':
            try:
                res2.append(group5_model.predict([inputs.iloc[i][f_group5]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_6':
            try:
                res2.append(group6_model.predict([inputs.iloc[i][f_group6]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_7':
            try:
                res2.append(group7_model.predict([inputs.iloc[i][f_group7]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_8':
            try:
                res2.append(group8_model.predict([inputs.iloc[i][f_group8]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_9':
            res2.append('LIHC')
        elif testtemp == 'group_10':
            try:
                res2.append(group10_model.predict([inputs.iloc[i][f_group10]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_11':
            try:
                res2.append(group11_model.predict([inputs.iloc[i][f_group11]])[0])
            except ValueError:
                res2.append('error')
        elif testtemp == 'group_12':
            try:
                res2.append(group12_model.predict([inputs.iloc[i][f_group12]])[0])
            except ValueError:
                res2.append('error')

    true = np.sum(np.array(res2) == np.array(test))
    print('The number of correct outcome:', true)
    print('The number of error outcome', len(test) - true)
    print('{:.2%}'.format(true / l) + ' ' + str(true) + '  ' + str(l))
    counti = list(np.array(res2) == np.array(test))
    for i in range(l):
        if counti[i] == False:
            print('The error outcome: ' + str(res2[i]) + ' | its correct result: ' + str(test[i]))

            
def PredictRank(inputs):#predict top result 
    test = list(inputs.index)
    l = len(test)
    for i in range(l):
        test[i] = test[i].split('.')[0]
    res2 = []
    label=firstlayer_model.classes_.tolist()
    la=len(label)
    error=['NA']
    OV=['OV']
    LIHC=['LIHC']
    THYM=['THYM']
    for i in range(l):
        prob=firstlayer_model.predict_proba([inputs.iloc[i][f_firstlayer]]).tolist()[0]
        dic={}
        for ii in range(la):
            dic[label[ii]]=prob[ii]
        a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
        testtemp=list(dict(a1))[0:2]
        tempres=[]    
        for j in testtemp:
            if j == 'group_1':
                tempres.append(OV)
            elif j == 'group_2':
                try:
                    prob=group2_model.predict_proba([inputs.iloc[i][f_group2]]).tolist()[0]
                    dic={}
                    label2=group2_model.classes_.tolist()
                    la2=len(label2)
                    for k in range(la2):
                        dic[label2[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append([error])
            elif j == 'group_3':
                tempres.append(THYM)
            elif j == 'group_4':
                try:
                    prob=group4_model.predict_proba([inputs.iloc[i][f_group4]]).tolist()[0]
                    dic={}
                    label4=group4_model.classes_.tolist()
                    la4=len(label4)
                    for k in range(la4):
                        dic[label4[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_5':
                try:
                    prob=group5_model.predict_proba([inputs.iloc[i][f_group5]]).tolist()[0]
                    dic={}
                    label5=group5_model.classes_.tolist()
                    la5=len(label5)
                    for k in range(la5):
                        dic[label5[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_6':
                try:
                    prob=group6_model.predict_proba([inputs.iloc[i][f_group6]]).tolist()[0]
                    dic={}
                    label6=group6_model.classes_.tolist()
                    la6=len(label6)
                    for k in range(la6):
                        dic[label6[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_7':
                try:
                    prob=group7_model.predict_proba([inputs.iloc[i][f_group7]]).tolist()[0]
                    dic={}
                    label7=group7_model.classes_.tolist()
                    la7=len(label7)
                    for k in range(la7):
                        dic[label7[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_8':
                try:
                    prob=group8_model.predict_proba([inputs.iloc[i][f_group8]]).tolist()[0]
                    dic={}
                    label8=group8_model.classes_.tolist()
                    la8=len(label8)
                    for k in range(la8):
                        dic[label8[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_9':
                tempres.append(LIHC)
            elif j == 'group_10':
                try:
                    prob=group10_model.predict_proba([inputs.iloc[i][f_group10]]).tolist()[0]
                    dic={}
                    label10=group10_model.classes_.tolist()
                    la10=len(label10)
                    for k in range(la10):
                        dic[label10[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_11':
                try:
                    prob=group11_model.predict_proba([inputs.iloc[i][f_group11]]).tolist()[0]
                    dic={}
                    label11=group11_model.classes_.tolist()
                    la11=len(label11)
                    for k in range(la11):
                        dic[label11[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)
            elif j == 'group_12':
                try:
                    prob=group12_model.predict_proba([inputs.iloc[i][f_group12]]).tolist()[0]
                    dic={}
                    label12=group12_model.classes_.tolist()
                    la12=len(label12)
                    for k in range(la12):
                        dic[label12[k]]=prob[k]
                    a1 = sorted(dic.items(), key=lambda x: x[1],reverse=True)
                    tempres.append(list(dict(a1))[0:2])
                except ValueError:
                    tempres.append(error)

        tempres = [item for sublist in tempres for item in sublist]
        res2.append(tempres)
    for i in range(len(res2)):
        if 'NA' in res2[i]:
            res2[i].remove('NA')
        print("The prediction results of "+str(test[i])+":"+str(res2[i]))
```
